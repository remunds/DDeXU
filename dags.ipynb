```python
import daft
import matplotlib.pyplot as plt

pgm = daft.PGM(aspect=1, dpi=200)
# leaves
for i in range(3):
    pgm.add_node(f"x_{i}", r"x", i*2+0.0, 0.0)
    pgm.add_node(f"y_{i}", r"y", i*2+0.6, 0.0)
    pgm.add_node(f"z_{i}", r"z", i*2+1.2, 0.0)

# sum nodes
for i in range(2):
    pgm.add_node(f"m_{i}", r"+", i*2+1.5, 1.5)

# factor node
pgm.add_node(f"f", r"*", 2.5, 2.5)

# connect all leaves to the inner nodes
for i in range(3):
    pgm.add_edge(f"x_{i}", f"m_0", directed=False, label=f"w_1")
    pgm.add_edge(f"y_{i}", f"m_0", directed=False, label=f"w_2") 
    pgm.add_edge(f"z_{i}", f"m_0", directed=False , label=f"w_3")
    pgm.add_edge(f"x_{i}", f"m_1", directed=False, label=f"w_4")
    pgm.add_edge(f"y_{i}", f"m_1", directed=False, label=f"w_5")
    pgm.add_edge(f"z_{i}", f"m_1", directed=False, label=f"w_6")
    

# connect the sum nodes to the factor node
pgm.add_edge(f"m_0", f"f", directed=False)
pgm.add_edge(f"m_1", f"f", directed=False)

pgm.render()
```
